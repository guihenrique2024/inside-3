```python
import pandas as pd
import plotly.express as px
import random

dates = pd.to_datetime(pd.date_range(start='2023-01-01', periods=100, freq='D'))
products = ['Tênis Air Max', 'Camiseta Dri-FIT', 'Shorts Esportivo', 'Jaqueta Corta Vento', 'Meia Cano Curto']

nike_data = {
    'Data': random.sample(list(dates), 50),
    'Produto': [random.choice(products) for _ in range(50)],
    'Valor': [round(random.uniform(50, 500), 2) for _ in range(50)]
}

df_nike = pd.DataFrame(nike_data)
df_nike = df_nike.sort_values(by='Data').reset_index(drop=True)

df_nike['Dia'] = df_nike['Data'].dt.day
df_nike['Mês'] = df_nike['Data'].dt.month
df_nike['Ano'] = df_nike['Data'].dt.year
df_nike['valor'] = df_nike['data'].dt.price

display(df_nike.head())

print("Tabela Nike da criada com sucesso!")
```

```python

fig_time = px.line(df_nike, x='Data', y='Valor', title='Vendas da Nike ao Longo do Tempo')
fig_time.show()


fig_product = px.bar(df_nike.groupby('Produto')['Valor'].sum().reset_index(),
                     x='Produto', y='Valor', title='Total de Vendas da Nike por Produto')
fig_product.show()
```
