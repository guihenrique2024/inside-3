```python
import pandas as pd
import plotly.express as px

# Criando um DataFrame de exemplo
data = {
    'Data': pd.to_datetime(['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05']),
    'Vendas': [100, 120, 90, 150, 110],
    'Custos': [50, 60, 45, 70, 55]
}
df = pd.DataFrame(data)

# Exibindo as primeiras linhas do DataFrame
display(df.head())
```

```python
# Criando um gráfico de linhas interativo com Plotly
fig = px.line(df, x='Data', y='Vendas', title='Vendas Diárias Interativas')
fig.show()
```
